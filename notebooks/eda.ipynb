### 📊 EDA Summary – Telco Customer Churn Dataset

- **Dataset Shape**: 7043 entries × 21 columns
- **Goal**: Predict customer churn (Yes/No) based on demographics, service usage, and billing data.

---

#### 🧾 Target Variable (Churn)
- Churned: **26.5%**, Not Churned: **73.5%** → Dataset is **imbalanced**
- Indicates a significant portion of customers leave early — worth modeling.

---

#### 📦 Key Numeric Insights
- **MonthlyCharges**:
  - Higher for churned customers.
  - Median charges are noticeably higher for churners than non-churners.
- **Tenure**:
  - Strong churn predictor.
  - Most churned customers have **tenure < 15 months**, while long-tenure customers are more stable.
- **TotalCharges**:
  - Correlates well with tenure (r = 0.83)
  - Requires conversion from object to numeric (some missing values).

---

#### 📊 Categorical Features
- **Contract Type**:
  - **Month-to-month** customers show the highest churn.
  - **Two-year** contracts have the lowest churn rates — strong negative indicator.
- **Add-on Services** like `TechSupport`, `OnlineSecurity` show visible patterns, potentially reducing churn (to be confirmed in feature importance later).

---

#### 🔗 Correlation Matrix
- **Tenure**, **TotalCharges**, and **MonthlyCharges** show moderate-to-strong positive correlations.
- **SeniorCitizen** has very low correlation with other features — low predictive power alone.

---

#### 🔍 Next Steps
- Convert categorical variables using encoding
- Handle missing `TotalCharges` rows
- Engineer new features if needed (e.g., `charge_ratio = MonthlyCharges / TotalCharges`)
- Prepare data for ML training

---

This EDA indicates several strong predictors of churn and shows clear differences in usage and contract behaviors between churned and retained customers.
